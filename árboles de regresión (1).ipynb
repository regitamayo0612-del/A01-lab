## **Árbol de regresión para predicciones**

```python
import pandas as pd

datos = [
    {"genero": "H", "edad": 21, "estatura": 1.73},
    {"genero": "M", "edad": 19, "estatura": 1.62},
    {"genero": "H", "edad": 21, "estatura": 1.73},
    {"genero": "H", "edad": 19, "estatura": 1.69},
    {"genero": "H", "edad": 20, "estatura": 1.75},
    {"genero": "H", "edad": 21, "estatura": 1.80},
    {"genero": "H", "edad": 22, "estatura": 1.70},
    {"genero": "H", "edad": 21, "estatura": 1.75},
    {"genero": "H", "edad": 19, "estatura": 1.66},
    {"genero": "M", "edad": 21, "estatura": 1.64},
    {"genero": "M", "edad": 19, "estatura": 1.62},
    {"genero": "H", "edad": 21, "estatura": 1.85},
    {"genero": "H", "edad": 20, "estatura": 1.75},
    {"genero": "H", "edad": 20, "estatura": 1.66},
    {"genero": "H", "edad": 19, "estatura": 1.70},
    {"genero": "H", "edad": 19, "estatura": 1.64}
]

df = pd.DataFrame(datos)

df_h_menor_21 = df[(df["genero"] == "H") & (df["edad"] < 21)]
print("Hombres menores a 21 años")
print(f"- Estatura promedio: {df_h_menor_21['estatura'].mean():.2f} m")

df_h_mayor_21 = df[(df["genero"] == "H") & (df["edad"] >= 21)]
print("Hombres de 21 años o más")
print(f"- Estatura promedio: {df_h_mayor_21['estatura'].mean():.2f} m")

df_m_19 = df[(df["genero"] == "M") & (df["edad"] == 19)]
print("Mujeres 19")
print(f"- Estatura promedio: {df_m_19['estatura'].mean():.2f} m")

df_m_21 = df[(df["genero"] == "M") & (df["edad"] == 21)]
print("Mujeres 21")
print(f"- Estatura promedio: {df_m_21['estatura'].mean():.2f} m")
```

### **Actividad**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

df = pd.read_csv('Advertising.csv')
df.head()
```

**Regresión lineal simple**

```python
X = df[['TV', 'radio', 'newspaper']]
y = df['sales']

modelo = LinearRegression()
modelo.fit(X, y)

m = modelo.coef_[0]      
b = modelo.intercept_    
print(f"Ecuación del modelo: Ventas = {m:.4f} * TV + {b:.4f}")
```

**Cálculo R2** 

```python
r2_lineal = modelo.score(X, y)

print(f"El R2 de la Regresión Lineal Simple es: {r2_lineal:.4f}")
```

**Árbol de Decisión para regresión**

```python
modelo_arbol = DecisionTreeRegressor(random_state=42)

modelo_arbol.fit(X, y)
```

```python
r2_arbol = modelo_arbol.score(X, y)

print(f"El R2 del Árbol de Decisión es: {r2_arbol:.4f}")
```

**Train-Test Split 50/50**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=42)

modelo_lin_split = LinearRegression()
modelo_lin_split.fit(X_train, y_train) 
r2_lin_split = modelo_lin_split.score(X_test, y_test) 

modelo_arbol_split = DecisionTreeRegressor(random_state=42)
modelo_arbol_split.fit(X_train, y_train) 
r2_arbol_split = modelo_arbol_split.score(X_test, y_test) 

print("Resultados Train/Test Split 50/50")
print(f"R2 Regresión Lineal en Prueba: {r2_lin_split:.4f}")
print(f"R2 Árbol de Decisión en Prueba: {r2_arbol_split:.4f}")
```

Incluso frente a datos desconocidos (Test), el Árbol de Decisión demostró ser superior para este caso específico, obteniendo un R² de 0.9345 frente al 0.8721 de la Regresión Lineal. Esto significa que el árbol logró descifrar un 93.4% del comportamiento de las ventas, capturando interacciones más complejas entre la inversión en TV, radio y periódico que una simple línea recta no pudo detectar.
