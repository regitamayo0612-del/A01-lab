```python
import pandas as pd
import numpy as np

data=pd.read_excel("Boston Housing Dataset 1978.xlsx")
data.head()
```

Escalamiento minmax

```python
datascaled=data.copy()
for col in datascaled.columns:
    minimo = datascaled[col].min()
    datascaled[col] = (datascaled[col]-minimo)
    maximo = datascaled[col].max()
    datascaled[col]/=maximo
```

```python
datascaled.describe()
```

Escalamiento estándar

```python
datastd = data.copy()

for col in datastd.columns:
    media = datastd[col].mean()
    std = datastd[col].std()
    
    datastd[col] = (datastd[col] - media) / std
```
