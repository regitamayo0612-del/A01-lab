```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

df=pd.read_excel("lab sabores agua.xlsx")
df.head()
```

```python

df['tamaño_bin_neg'] = df['tamaño'].map({'grande': -1, 'chica': 1})
df['hielo_bin_neg'] = df['hielo'].map({'con': 1, 'sin': -1})

df['kahlua_bin'] = df['sabor'].map({'kahlúa': 1, 'horchata': 0})
df['horchata_bin'] = df['sabor'].map({'kahlúa': 0, 'horchata': 1})

df['tamaño'] = df['tamaño'].map({'grande': 1, 'chica': 0})
df['hielo'] = df['hielo'].map({'con': 1, 'sin': 0})

df = pd.get_dummies(df, columns=['sabor'], prefix='', prefix_sep='', dtype=int)
print(df.head())
```

**Modelo 1 usando 0 y 1**

```python
X = np.array(df[['tamaño', 'hielo', 'kahlúa','horchata']])
y = np.array(df['score'])

lr = LinearRegression()
lr.fit(X, y)
print("Modelo 1: Sin Penalización")
print("beta_0 (Intercepto):", lr.intercept_)
print("Coeficientes (beta_1, beta_2, beta_3):", lr.coef_)
print("R^2 score:", lr.score(X, y))
```

```python
x2=sm.add_constant(X)
model=sm.OLS(y,x2)
results=model.fit()
print(results.summary())
```

**Modelo 2 usando -1 y 1**

```python
X2 = df[['tamaño_bin_neg', 'hielo_bin_neg', 'kahlua_bin','horchata_bin']]

lr2 = LinearRegression()
lr2.fit(X2, y)
print("beta_0 (Intercepto):", lr2.intercept_)
print("Coeficientes (beta_1, beta_2, beta_3):", lr2.coef_)
print("R^2 score:", lr2.score(X2, y))
```

```python
x3=sm.add_constant(X2)
model=sm.OLS(y,x3)
results=model.fit()
print(results.summary())
```

Mientras que el Modelo 1 (0, 1) facilita identificar el valor de partida o la línea base y el valor añadido de cada variable, el Modelo 2 (-1, 1) es superior para identificar qué tanto se aleja cada atributo del promedio general
