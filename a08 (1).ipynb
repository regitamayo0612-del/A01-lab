## **A08 - Bootstrapping + Aggregating**

## **Bootstrap**

**Dataset**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
df.head()
```

```python
X = df[['hp', 'qsec']]
X = sm.add_constant(X)
y = df['mpg']
```

**2) Regresión lineal**

```python
model = sm.OLS(y, X).fit()
print(model.summary())
```

**2a) Intervalos de confianza**

```python
print(model.conf_int())
```

**3) Bootstrap**

```python
n = len(df)
B = 1000 
betas = []
```

```python
for i in range(B):
    # bootstrap con reemplazo
    sample = df.sample(n=n, replace=True)
    
    X = sample[['hp', 'qsec']]
    X = sm.add_constant(X)
    y = sample['mpg']
    
    model = sm.OLS(y, X).fit()
    
    betas.append(model.params.values)
betas = pd.DataFrame(betas, columns=['beta0', 'beta1_hp', 'beta2_qsec'])
```

```python
beta_media = betas.mean()
beta_std = betas.std()
beta_ci = betas.quantile([0.025, 0.975])
```

```python
print("Media de betas:")
print(beta_media)

print("Desviación estándar:")
print(beta_std)

print("Intervalos de confianza :")
print(beta_ci)
```

```python
lower = beta_media - 2 * beta_std
upper = beta_media + 2 * beta_std

for i, name in enumerate(['b0', 'b1 (hp)', 'b2 (qsec)']):
    print(f"{name}: [{lower[i]:.4f}, {upper[i]:.4f}]")
```

## **Aggregating**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


y = df['mpg']
X = df.drop(columns=['mpg', 'model'])
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.5, random_state=42)
```

**Función Train**

```python
def train_ensemble(X_train, y_train, n_models=1000, n_features=3):
    
    ensemble = []
    
    for i in range(n_models):
        
        vars_i = np.random.choice(X_train.columns, size=n_features, replace=False)
        X_train_i = sm.add_constant(X_train[vars_i])
        
        model = sm.OLS(y_train, X_train_i).fit()
        
        ensemble.append({
            "model": model,
            "vars": vars_i
        })
    
    return ensemble
```

**Función Test**

```python
def predict_ensemble(ensemble, X_new):
    
    preds = []
    
    for item in ensemble:
        model = item["model"]
        vars_i = item["vars"]
        
        X_new_i = sm.add_constant(X_new[vars_i], has_constant='add')
        
        y_hat = model.predict(X_new_i)
        preds.append(y_hat.values)
    
    preds = np.array(preds)
    
    y_hat_mean = preds.mean(axis=0)
    y_hat_std = preds.std(axis=0)
    
    return y_hat_mean, y_hat_std
```

Entrenar

```python
ensemble = train_ensemble(X_train, y_train)
```

Predecir en test

```python
y_hat_test, y_hat_std = predict_ensemble(ensemble, X_test)
```

```python
r2 = r2_score(y_test, y_hat_test)

print("R2 en test:", r2)
```

## **Random Forest**

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score

df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
X = df.drop(columns=['mpg', 'model'])
y = df['mpg']
```

Configurar el modelo optimizado

```python
rf_final = RandomForestRegressor(
    n_estimators=190, 
    max_features='sqrt', 
    random_state=42
)

kf = KFold(n_splits=10, shuffle=True, random_state=42)
```

Validación cruzada para R2

```python
r2_scores = cross_val_score(rf_final, X, y, cv=kf, scoring='r2')

print(f"R2 promedio: {np.mean(r2_scores):.4f}")
print(f"Desviación estándar: {np.std(r2_scores):.4f}")
print(f"Scores por fold:\n{r2_scores}")
```

Con GridSearchCV

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, GridSearchCV

df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
X = df.drop(columns=['mpg', 'model'])
y = df['mpg']

# Configurar el modelo y la rejilla de parámetros
rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200, 300, 500], 
    'max_depth': [None, 2, 3, 4, 5],           
    'max_leaf_nodes': [None, 5, 10, 15, 20]   
}

# Configurar k-folds
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Grid Search
grid_search = GridSearchCV(
    estimator=rf, 
    param_grid=param_grid, 
    cv=kf, 
    scoring='r2', 
    n_jobs=-1
)
grid_search.fit(X, y)


print(f"Mejor R2 encontrado: {grid_search.best_score_:.4f}")
print(f"Mejores parámetros: {grid_search.best_params_}")
```

**Concluyo que...** Aunque el primer modelo alcanzó un $R^2$ promedio superior (0.5853), su alta desviación estándar de 0.3722 indica que el resultado es inestable y depende en gran medida de cómo se dividieron los datos, llegando incluso a tener predicciones con precisión negativa en ciertos grupos. Por el contrario, la búsqueda de rejilla identificó que una estructura más controlada, con una profundidad máxima de 4 y un bosque de 500 árboles, es la configuración más robusta para este conjunto de datos, a pesar de reportar un $R^2$ de 0.5093.

**Boosting**

el Boosting construye árboles de forma secuencial: cada nuevo árbol intenta corregir los errores cometidos por los anteriores

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold, GridSearchCV

df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
X = df.drop(columns=['mpg', 'model'])
y = df['mpg']

kf = KFold(n_splits=10, shuffle=True, random_state=42)

gb = GradientBoostingRegressor(random_state=42)
```

Rejilla de parámetros

```python
param_grid = {
    'n_estimators': [50, 100, 200, 500],
    'max_depth': [2, 3, 4, 5],
    'max_leaf_nodes': [None, 5, 10, 15],
    'learning_rate': [0.01, 0.05, 0.1]
}
```

Ejecutar Grid Search

```python
grid_search_gb = GridSearchCV(
    estimator=gb,
    param_grid=param_grid,
    cv=kf,
    scoring='r2',
    n_jobs=-1
)
grid_search_gb.fit(X, y)

print(f"Mejor R2 con Boosting: {grid_search_gb.best_score_:.4f}")
print(f"Mejores Parámetros: {grid_search_gb.best_params_}")
```
